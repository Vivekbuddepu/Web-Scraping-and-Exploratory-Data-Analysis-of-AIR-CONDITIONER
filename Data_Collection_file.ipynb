```python
import numpy as np
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

#### Extracting information from all the url using loops 

```python
brand=[]
capacity = []
star_Rating = []
Type = []
Ratings = []
Annual_power = []
warranty = []
room_size = []
price_after_discount = []
discount_percentage = []
Model_year = []
original_price = []
for i in range(2,47):
    url="https://www.flipkart.com/search?q=ac&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="+str(i)
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    b = soup.find_all('div',class_="KzDlHZ")
   
    for i in b:
        brand.append(re.findall(r"\w+\s\w+",i.text)[0])
    
    
    for i in b:
        a = re.findall(r"\d{4}|[0-9]{4}",i.text.split("Model")[0])
        if len(a) > 0:
            Model_year.append(a[0])
        else:
            Model_year.append(np.nan)
    
    
   
    for i in b:
        capacity.append(re.findall(r"\d+\.\d+\sTon|\d\sTon",i.text))
    
    

    for i in b:
        if len(re.findall(r"\d\sStar",i.text)) >0:
            star_Rating.append(re.findall(r"\d\sStar",i.text)[0])
        else:
            star_Rating.append(np.nan)
    
    

    for i in b:
        a = re.findall(r"Star\s\w+\s\w+",i.text)
        if len(a) >=2:
            Type.append(re.findall(r"Star\s\w+\s\w+",i.text)[1])
        elif len(a) == 1:
            Type.append(re.findall(r"Star\s\w+\s\w+",i.text)[0])
        else:
            Type.append(np.nan)
    
    
    r = soup.find_all('div',"XQDdHH")
    for i in r[:24]:
        Ratings.append(i.text)
    
    p = soup.find_all('li','J+igdf')

    for i in p:
        if "Annual" in i.text:
            a = re.findall(r"Annual Power Usage:.*",i.text)
            if len(a) > 0:
                Annual_power.append(a[0].split(":")[1].strip())
            else:
                Annual_power.append(np.nan)
        elif "Room Size" in i.text:
            b = re.findall(r"\bRoom Size:.*|room size:.*|Room.*",i.text)
            if len(b) > 0:
                room_size.append(b[0].split(":")[1].split("sqft")[0].strip())
            else:
                room_size.append(np.nan)
    
        else:
            c = re.findall(r"\W\d\s\w+\sStandard|\d+\s\w+\s\w+\s\w+\sWhole\sProduct|\d\s\w+\s\w+\s\w+\sProduct|\d\s\w+\s\w+\s\w+\s\w+\s[pP]roduct|\d\s\w+\s\w+\s\w+\s\w+\sAC|\d\s\w+\s\w+\sProduct",i.text)
            if len(c) > 0:
                warranty.append(c[0])
            else:
                warranty.append(np.nan)

    
    v = soup.find_all('div',"Nx9bqj _4b5DiR")
    
    for i in v:
        price_after_discount.append(i.text)

    
    i = soup.find_all('div',"yRaY8j ZYYwLA")
    
    for j in i:
        if "₹" in  j.text:
            original_price.append(j.text)
        else:
            original_price.append(np.nan)
    
    d = soup.find_all('div',"UkUFwK")
    
    for i in d[:24]:
        discount_percentage.append(i.text.split("%")[0])
```
